```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Параметры
e = 0 # Среднее значение для оси x
S = 5 # Дисперсия для оси x
# Генерация данных
data = np.random.multivariate_normal([e, 0], [[S, 2], [2, 211]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
# Гистограммы
for col in 'xy':
  plt.hist(data[col], density=True, alpha=0.5, label=f'{col}-axis')
  plt.legend()
  plt.title("Гистограммы по осям x и y")
  plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Генерация данных
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 211]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
# Построение графика с корректным синтаксисом
with sns.axes_style('white'):
  sns.jointplot(x="x", y="y", data=data, kind='kde')
  plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the planets dataset
planets = sns.load_dataset('planets')

with sns.axes_style('white'):
    g = sns.catplot(x="year", data=planets, aspect=2,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=5)
    plt.show()
    #Этот код Python использует библиотеки Seaborn и Matplotlib для генерации и отображения графика количества, визуализирующего количество планет, открываемых каждый год в наборе данных «планеты».
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the planets dataset
planets = sns.load_dataset('planets')

with sns.axes_style('white'):
    g = sns.catplot(x="year", data=planets, aspect=4.0, kind='count',
                       hue='method', order=range(2001, 2015))
    g.set_ylabels('Number of Planets Discovered')
    plt.show()
# код Python генерирует график категориального количества с использованием библиотек Seaborn и Matplotlib, визуализируя количество планет, открытых каждый год между 2001 и 2014 годами, категоризированных по методу открытия.
```
