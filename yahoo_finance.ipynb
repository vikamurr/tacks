```python
import yfinance as yf
import matplotlib.pyplot as plt

# Создание объекта Ticker для заданной компании
AMZN = yf.Ticker("AMZN")

# Получение цены закрытия акций за последний год
hist = AMZN.history(period="1y")

# Рассчет скользящего среднего за 30 дней
hist["SMA_90"] = hist["Close"].rolling(window=30).mean()

# Вывод графика цены закрытия акций и скользящего среднего за последний год
plt.plot(hist["Close"])
plt.plot(hist["SMA_90"])
plt.legend(["Close", "SMA_90"])
plt.show()
```

```python
import yfinance as yf
import pandas as pd

# Создание объектов Ticker для двух компаний
nke = yf.Ticker("NKE")
amzn = yf.Ticker("AMZN")

# Получение цены закрытия акций за последний год
nke_hist = nke.history(period="1y")
amzn_hist = amzn.history(period="1y")  # Исправлено

# Объединение данных в одну таблицу
data = pd.concat([nke_hist["Close"], amzn_hist["Close"]], axis=1)
data.columns = ["NKE", "AMZN"]

# Рассчет корреляции между ценами акций двух компаний
corr = data.corr()

# Вывод матрицы корреляции
print(corr)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объекта Ticker для заданной компании
aapl = yf.Ticker("AMZN")

# Получение цены закрытия акций за последние 5 лет
hist = aapl.history(period="5y")

# Рассчет процентного изменения цены закрытия акций
hist["Pct_Change"] = hist["Close"].pct_change()

# Вывод графика процентного изменения цены закрытия акций
plt.plot(hist["Pct_Change"])
plt.title("AMZN Percent Change")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объекта Ticker для заданной компании
nke = yf.Ticker("NKE")

# Получение цены закрытия акций за последние 5 лет
hist = nke.history(period="5y")

# Рассчет волатильности цены закрытия акций за 30 дней
hist["Volatility"] = hist["Close"].rolling(window=30).std()

# Вывод графика волатильности цены закрытия акций
plt.plot(hist["Volatility"])
plt.title("NKE Volatility")
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Создание объектов Ticker для заданных компаний
nke = yf.Ticker("NKE")
aapl = yf.Ticker("AMZN")

# Получение цены закрытия акций за последние 5 лет
nke_hist = nke.history(period="5y")
aapl_hist = aapl.history(period="5y")

# Объединение таблиц цен закрытия акций для обеих компаний
merged = pd.merge(nke_hist["Close"], aapl_hist["Close"], left_index=True, right_index=True, suffixes=("_NKE", "_AMZN"))

# Вывод графиков цен закрытия акций обеих компаний
merged.plot()
plt.title("NKE vs AMZN Close Prices")
plt.show()

# Рассчет коэффициента корреляции между ценами закрытия акций обеих компаний
corr = merged.corr().iloc[0,1]

# Вывод коэффициента корреляции
print("NKE vs AMZN Correlation: {:.2f}".format(corr))
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Задание тикеров двух компаний
ticker1 = "NKE"
ticker2 = "AMZN"

# Получение цен закрытия за заданный период времени для двух компаний
start_date = "2025-01-01"
end_date = "2025-02-25"
stock1 = yf.Ticker(ticker1)
stock2 = yf.Ticker(ticker2)
stock1_hist = stock1.history(start=start_date, end=end_date)
stock2_hist = stock2.history(start=start_date, end=end_date)

# Создание DataFrame из цен закрытия для двух компаний
df = pd.concat([stock1_hist["Close"], stock2_hist["Close"]], axis=1)
df.columns = [ticker1, ticker2]

# Рассчет дневных доходностей для двух компаний
df_returns = df.pct_change()

# Рассчет корреляции между доходностями двух компаний
corr = df_returns.corr().iloc[0,1]

# Построение scatter plot
plt.scatter(df_returns[ticker1], df_returns[ticker2])
plt.title(f"{ticker1} vs {ticker2} - Correlation: {corr:.2f}")
plt.xlabel(f"{ticker1} Returns")
plt.ylabel(f"{ticker2} Returns")
plt.show()
```
